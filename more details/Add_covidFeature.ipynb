```python
sum= covid19[(covid19.Records == "Confirmed")].groupby(['Date'])['RecordsNo'].sum()
pre_confirmed=sum.to_dict()
sum
```

```python
def deal_date(date):
    oldDate = date.split('/')
    if len(oldDate[0]) < 2:
        oldDate[0]= '0' + oldDate[0]
    if len(oldDate[1]) < 2:
        oldDate[1]= '0' + oldDate[1]
    newDate = '/'.join(oldDate)
    return newDate
```

```python
confirmed=dict()
for i in pre_confirmed.items():
    #print(i)
    confirmed[deal_date(i[0])]=i[1]
confirmed = dict(sorted(confirmed.items()))
confirmed
```

```python
with open('/Users/yangtang/Desktop/Mar_TwentyFifth_reliability-11000.json') as f:
    data = json.load(f, strict=False)
```

```python
import datetime 
for i in range(len(data)):
    time=data[i]['create_at']
    cr_date = datetime.datetime.strptime(time[:10], '%Y-%m-%d')
    cr_date = cr_date.strftime("%m/%d/%Y")
    cr_date=cr_date[:-4]+cr_date[-2:]
    date=cr_date
    data[i]['confirmed_number']=confirmed[date]
    data[i]['date']=date
```
